# Linear Discriminant Analysis

---



**In order to find the number of componenets required for LDA, we can use a simple formula of n-1**

**Assume there are n persons, so how many lined we would need to distinguish between the given n persons? It will be n-1**

**In case you are having less number of features as compared to your classes, you need to use min(no_of_features, no_of_classes-1)**

```python
# If we have 19 classes and there are only 8 features
min(8,19-1)
# 8 will be the n_components
```

**In your data, if you have suppose 64 features, and number of classes are 10**

```python
# min(64,10-1)
# n_components = 10-1 => 9
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,train_size=0.8,random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression(max_iter=200)
```

```python
logistic_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
X_lda = lda.fit_transform(X,Y)
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X_lda)
```

```python
x_train_lda , x_test_lda , y_train_lda , y_test_lda = train_test_split(X_lda, Y, test_size=0.2,random_state=5)
```

```python
logistic_lda = LogisticRegression(max_iter=200)
```

```python
logistic_lda.fit(x_train_lda,y_train_lda)
```

```python
y_pred_lda = logistic_lda.predict(x_test_lda)
```

```python
accuracy_score(y_test_lda,y_pred_lda)
```
